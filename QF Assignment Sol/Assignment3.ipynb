Let's start by computing the mean-variance optimal portfolio across these five factors

I would like you start by using the full sample to estimate the weights and the leverage up the weights so this portfolio has the same in sample volatiltiy as the market portfolio

````python
df_factors = df.copy()
df_factors.rename(columns = {"Mkt-RF":"Mkt"}, inplace=True)
df_factors.drop(["RF"], axis=1, inplace=True)
df_factors.head()
````

````python
ER = df_factors.mean(axis=0)
CovR = df_factors.cov()
CovR_inv = np.linalg.inv(CovR)

# MVE
W = pd.DataFrame([], index = df_factors.columns)
w_mve = ER @ CovR_inv
W["MVE"] = w_mve

# Scaling
W["MVE"] = w_mve * df_factors["Mkt"].std(axis=0) / (ER @ CovR_inv @ ER)**0.5
print(W)

# Checking the volatility
print(f'The market factor volatility is {df_factors["Mkt"].std(axis=0)}')

# Portfolio volatility
mve_ret = df_factors @ W["MVE"]
mve_std = mve_ret.std()
print(f'The mve portfolio volatility is {mve_std}')

# Mean and Sharpe
print(f'The average return of the MVE portfolio is {mve_ret.mean()}')
print(f'The sharpe ratio of the MVE portfolio is {mve_ret.mean()/mve_std}')
      
````

Meaning that instead of estimating expected returns you simply use the vector of standard deviations $c[\sigma(R^1_t),\sigma(R^2_t),\sigma(R^3_t),..]$. 4. Risk Parity 2: Assume that all factors have zero correlation with each other

Lets construct market hedged factors!

Say for each factor, you have the returns $R^{factor}_t$ in a dataframe. You also have it's beta $\beta$, and the returns on the market portfolio (the excess return to be precise) $R^{mkt-rf}_t$


So you can construct the market hedged as


$$R^{factor-market}_t=R^{factor}_t-\beta*R^{mkt-rf}_t$$

Compute this for all other four factors


Report means, standard deviations, and also the beta with respect to the market





````python
Hedged = pd.DataFrame(index = df_factors.index, columns = df_factors.columns)

for factor in df_factors.columns:
    Hedged[factor] = df_factors[factor].subtract(betas.loc["beta", factor]*df["Mkt-RF"], axis=0)


Hedged.head()
````
