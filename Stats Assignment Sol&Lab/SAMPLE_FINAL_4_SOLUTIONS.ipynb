```python
import pandas as pd  
import numpy as np  
import statsmodels.formula.api as smf
```

```python
dataset = pd.read_csv('http://paulellickson.com/ClassData/mortgage.csv')
dataset.head()
```

```python
est1 = smf.ols(formula='deny ~ pi_rat + single + selfemp + condo + probunmp', data=dataset).fit()
print(est1.summary())
```

```python
est2 = smf.ols(formula='deny ~ pi_rat + single + selfemp', data=dataset).fit()
print(est2.summary())
```

```python
print(est1.f_test("condo = 0,probunmp = 0"))
```

```python
d = {'pi_rat': [.4], 'single': [1], 'selfemp': [1]}
df = pd.DataFrame(data=d)
df
```

```python
predictions = est2.get_prediction(df)
predictions.summary_frame(alpha=0.05)
```

```python
d = {'pi_rat': [.4], 'single': [0], 'selfemp': [0]}
df = pd.DataFrame(data=d)
df
```

```python
predictions = est2.get_prediction(df)
predictions.summary_frame(alpha=0.05)
```
