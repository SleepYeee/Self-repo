```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
```

```python
dataset = pd.read_csv('http://paulellickson.com/ClassData/credapp.csv')
dataset.head()
```

```python
est1 = smf.ols(formula='approved ~ age + ataddress + dependents + badcred1 + badcred2 + ownshome +income + selfempl +incper', data=dataset).fit()
print(est1.summary())
```

```python
est2 = smf.ols(formula='approved ~ badcred1 + badcred2 + ownshome +income', data=dataset).fit()
print(est2.summary())
```

```python
print(est1.f_test("age = 0,ataddress = 0,dependents=0,selfempl=0,incper=0"))
```

```python
d = {'badcred1': [1], 'badcred2': [1], 'ownshome': [1], 'income': [5000]}
df = pd.DataFrame(data=d)
df
```

```python
predictions = est2.get_prediction(df)
predictions.summary_frame(alpha=0.05)
```
