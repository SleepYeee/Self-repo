```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
```

```python
dataset = pd.read_csv('http://paulellickson.com/ClassData/credit.csv')
dataset.head()
```

```python
est1 = smf.ols(formula='Rating ~ Income + Cards + Age + Education + Male + Student + Married', data=dataset).fit()
print(est1.summary())
```

```python
est2 = smf.ols(formula='Rating ~ Income + Cards + Age', data=dataset).fit()
print(est2.summary())
```

```python
print(est1.f_test("Education = 0,Male = 0,Student=0,Married=0"))
```

```python
d = {'Income': [125], 'Cards': [2], 'Age': [40]}
df = pd.DataFrame(data=d)
df
```

```python
predictions = est2.get_prediction(df)
predictions.summary_frame(alpha=0.10)
```

```python
est2 = smf.ols(formula='np.log(Rating) ~ Income + Cards + Age', data=dataset).fit()
print(est2.summary())
```
