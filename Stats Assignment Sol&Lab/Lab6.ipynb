

 **Case Study: Employee tenure and store performance (Similar to HW5)**

You run a family business that sell fresh farm goods and you have opened serveral stores in the local area. To boost the financial
performance, you want to have a better idea of the relationship of employee retention and store profit. A file named **lab_data.csv**
contains data you have regarding the store-level characteristics in the following 9 aspects: Profit, MTenure (on average, how
many months have a manager been with this store), CTenure (on average, how many months have a crew member been with this
store), Comp (number of competitiors per 100,000 people in the city), Pop (city population), Visiblity (visibility rating from 1 to
5, with 5 being the highest), PedCount (pedestrian foot traffic rating from 1 to 5, with 5 being the highest), Hours24 (indicator
equal to 1 if the store opens 24 hours), Res (indicator equal to 1 if the city is largely a residential, as opposed to industrial, area)



1.Run a regression of *Profit* on all the explanatory variables above. Based on your regression results, intepret the coefficient of *MTenure* and *CTenure* and evaluate their statistical significance at the 1% significance level. What about Res?



```python
# import packages
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from scipy import optimize
```

```python
lab_data = pd.read_csv("lab_data.csv")
lab_data.head()
```

basic data investigation and simple regressions

```python
lab_data.describe()
```

```python
reg=smf.ols(formula='Profit~MTenure+CTenure+Pop+Comp+Visibility+PedCount+Res+Hours24', data=lab_data)
est = reg.fit()
print(est.summary())
```



*   The coefficient of *MTenure* is roughly 5.5, implying that an additional month of manager tenure is associated with an expected increase in (yearly) profit of $5.5 holding other things constant; *MTenure *is statistically significant at the 1% level (p-value < 0.01);

*   The interpretation of *CTenure* is pretty similar to *MTenure*. *CTenure* is not significant at the 1% level. However, it is significant at the 10% level (0.05 < p-value < 0.1)

* The coefficient of *Res* implies that, compared with stores located at the industrial area, stores located at the residential area would have a $709 higher (yearly) profit holding other things constant. It is not significant at the 1% level, but significant at the 5% level (0.01 < p-value < 0.05)


2.Before the regression analysis, you were considering contracting part of managers’ salaries on their store crew retention (i.e., if

```python
change_profit = est.params['CTenure'] * 1.412
print("Expected profit change:" , change_profit)
```

5.How would you design the regression if the impact of *MTenure* and *CTenure* on *Profit* varies with the level of tenure? Run the regression and find the point where an additional month of MTenure leads to lower profit.

```python
form1='Profit~MTenure+np.power(MTenure,2)+CTenure+np.power(CTenure,2)+Pop+Comp+Visibility+PedCount+Res+Hours24'
est = smf.ols(formula=form1, data=lab_data).fit()
print(est.summary())
```

* We do find diminishing returns to MTenure and CTenure ! (any thoughts on why? What else do you observe from the regression output?)

* To find the point where the profit starts to decrease with MTenure , we just need to find the maximal point, because the profit first increases, then decreases.

Profit = 13.68 MTenure - 0.037 MTenure^2  + .... (terms unrelated to Mtenure)
- Take derivative with respect to *Mtenure*, 13.68 - 0.074MTenure = 0 $\rightarrow$ MTenure = 184 months

```python
#the other apporach

def f(x):
  return -(est.params['MTenure']*x + est.params['np.power(MTenure, 2)']*x**2)
result = optimize.minimize_scalar(f)
print(result.success) # check if solver was successful
```
