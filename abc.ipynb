```python
import yfinance as yf
import numpy as np
import pandas as pd
from datetime import datetime, date, timedelta
```

```python
start = datetime(2014, 1, 1)
end = datetime.today()

tckr = 'BTC-USD'
ticker = yf.Ticker(tckr)

df = ticker.history(start=start,
                    end=end,
                    interval="1d")
```
